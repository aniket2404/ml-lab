# [Objective 03](#pandas-visualizations) Describe and discriminate between basic data types

## Overview

One of the cornerstones of Exploratory Data Analysis (EDA) is being able to identify variable types.  We will need different statistical methods to display and describe each of these different types of data.

```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```
df= pd.read_csv('/content/iris_dirty.csv')
```

```

df.describe()
```

```
df.info()
```

```
df['Species'].value_counts()
```

```
df['Species'] = df['Species'].str.lower()
df['Species'].value_counts()
```

```
df.drop(columns=['Unnamed: 0'], inplace=True)
```

```
sns.set(style="dark")
plt.style.use("dark_background")

num_cols = df.select_dtypes(include=['float64', 'int64']).columns

plt.figure(figsize=(12, 8))

for i, col in enumerate(num_cols, 1):
    plt.subplot(2, 2, i)
    sns.histplot(df[col].dropna(), bins=10, color="red", stat="density")
    sns.kdeplot(df[col].dropna(), color="cyan", linewidth=2)

    plt.title(f"Distribution of {col}")
    plt.xlabel(col)
    plt.ylabel("Count")

plt.tight_layout()
plt.show()
```

```
plt.figure(figsize=(6, 4))
sns.barplot(data=df, x='Species', y='Petal.Length', estimator='mean', palette='coolwarm',hue="Species")
plt.title("Average Petal Length per Species")
plt.xlabel("Species")
plt.ylabel("Avg. Petal Length")
plt.tight_layout()
plt.show()
```

```
plt.figure(figsize=(12,8))
sns.set(style="white")
plt.style.use("seaborn-v0_8-whitegrid")
for i,col in enumerate(num_cols, 1):
    plt.subplot(2, 2, i)
    sns.boxplot(x="Species", y=col, data=df, palette="cubehelix", hue='Species')
    plt.title(f"{col} Distribution by Species")

plt.tight_layout()
plt.show()
```

```
plt.figure(figsize=(8,4))
sns.heatmap(df[num_cols].corr(), annot=True, cmap="Spectral", fmt=".1f", cbar=True)
plt.title("Correlation Heatmap", fontsize=12)
plt.tight_layout()
plt.show()
```

```
sns.countplot(data=df, x='Species', palette='Set2',hue="Species")
plt.title("Count of Each Species")
plt.show()
```

```
sns.pairplot(df, hue='Species', palette='husl', diag_kind='kde')
plt.suptitle("Pairwise Plots of Iris Features", y=1.02)
plt.show()
```
